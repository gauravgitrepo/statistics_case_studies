```python
### UDF For Continous Variable Summary

def Conti_Vars_Summary(x):
    return pd.Series([x.count(),x.isnull().sum(),x.sum(),x.mean(),x.median(),x.var(),x.std(),x.min(),x.quantile(0.01),
                     x.quantile(0.05),x.quantile(0.10),x.quantile(0.25),x.quantile(0.50),x.quantile(0.75),
                     x.quantile(0.90),x.quantile(0.95),x.quantile(0.99),x.max()],index=['N','NMISS','SUM','MEAN','MEDIAN','VAR',
                                                                                       'STD','MIN','P1','P5','P10','P25','P50','P75','P90',
                                                                                       'P95','P99','MAX'])
```

```python
### UDF For Categorical Variable Summary

def Catog_Vars_Summary(x):
    Mode= x.value_counts().sort_values(ascending=False)[0:1].reset_index()
    return pd.Series([x.count(),x.isnull().sum(),Mode.iloc[0,0],Mode.iloc[0,1],
                     round(Mode.iloc[0,1]*100/x.count(),2)],index=['N','NMISS','MODE','FREQ','PERCENT'])
```

```python
### UDF For Missing Value Imputation

def missing_value_imputation(x,stats='mean'):
    if (x.dtypes == 'int64') | (x.dtypes == 'float64'):
        x = x.fillna(x.mean() if stats == 'mean' else x.fillna(x.median())) 
    else:
        x = x.fillna(x.mode())
    return x
    
```
